```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from matplotlib.gridspec import GridSpec
import warnings

# --- Configuration ---
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
RANDOM_SEEDS = [42, 1, 2, 3]

# --- Step 1: Load Your Data ---
try:
    df_full = pd.read_csv('SP500_2021_2025.csv', parse_dates=['Date'], dayfirst=True)
    df_full.sort_values('Date', inplace=True)
    print("Successfully loaded 'SP500_2021_2025.csv'.")
except FileNotFoundError:
    raise SystemExit("Error: 'SP500_2021_2025.csv' not found. Please ensure the file is in the correct directory.")

# --- Step 2: Define Metrics from Paper ---

def calculate_ncdc(data, labels, k, centroids):
    """
    Calculates the Normalized Centroid Distance times Centroids (NCDC)
    based on the corrected formula from oldmain.tex.
    Formula: NCDC_k = k * D_k / D_1
    Where D_1 is the total distance of all points to the overall mean,
    and D_k is the sum of distances of points to their centroids, plus
    the sum of distances of all centroids to the overall mean.
    """
    if len(data) == 0:
        return np.inf

    mean_center = np.mean(data)
    d1 = np.sum(np.abs(data - mean_center))

    if d1 == 0:
        return np.inf

    dk = 0

    # Sum of distances from points to their own centroids
    for i in range(k):
        cluster_data = data[labels == i]
        if len(cluster_data) > 0:
            dk += np.sum(np.abs(cluster_data - centroids[i]))

    # Sum of distances from centroids to the overall mean center
    dk += np.sum(np.abs(centroids - mean_center))

    ncdc = k * dk / d1
    return ncdc

def get_sc_plus(sc_scores, ncdc_scores):
    """
    Applies the SC+ rule from the paper to determine the decisive metric.
    Rule: if max(SC) for K>1 > 0.65, use SC scores. Otherwise, use NCDC scores.
    """
    # Max SC is calculated for K > 1, as per the paper's definition
    max_sc_for_decision = max({k: v for k, v in sc_scores.items() if k > 1}.values()) if len(sc_scores) > 1 else -1

    if max_sc_for_decision > 0.65:
        return sc_scores, 'SC'
    else:
        return ncdc_scores, 'NCDC'

def find_optimal_k(scores, metric_name):
    """Finds optimal K based on metric (max for SC, min for NCDC)."""
    if not scores:
        return 1
    if metric_name == 'SC':
        # SC for K=1 is defined as 0, so we should only consider K>1 for finding the max.
        filtered_scores = {k: v for k, v in scores.items() if k > 1}
        return max(filtered_scores, key=filtered_scores.get) if filtered_scores else 1
    elif metric_name == 'NCDC':
        return min(scores, key=scores.get)
    return 1

# --- Step 3: Define the Core Analysis and Visualization Function ---
def perform_analysis_and_plot(data_to_cluster, year, random_seed, df_2025, analysis_type='SC+'):
    """
    Performs analysis and generates plots based on the specified methodology.
    """
    sc_scores, ncdc_scores = {}, {}
    k_range = range(1, 9)

    for k in k_range:
        if k == 1:
            sc_scores[k] = 0

            kmeans_k1 = KMeans(n_clusters=1, random_state=random_seed, n_init=10).fit(data_to_cluster)
            labels_k1 = np.zeros(len(data_to_cluster))
            ncdc_scores[k] = calculate_ncdc(data_to_cluster.flatten(), labels_k1, 1, kmeans_k1.cluster_centers_.flatten())
        else:
            if len(data_to_cluster) < k:
                continue
            kmeans = KMeans(n_clusters=k, random_state=random_seed, n_init=10).fit(data_to_cluster)
            labels = kmeans.labels_
            centroids = kmeans.cluster_centers_.flatten()
            sc_scores[k] = silhouette_score(data_to_cluster, labels)
            ncdc_scores[k] = calculate_ncdc(data_to_cluster.flatten(), labels, k, centroids)

    max_sc_for_decision = max({k: v for k, v in sc_scores.items() if k > 1}.values()) if len(sc_scores) > 1 else -1

    if analysis_type == 'SC+':
        if max_sc_for_decision > 0.65:
            decisive_scores = sc_scores
            metric_used = 'SC'
            optimal_k = find_optimal_k(sc_scores, 'SC')
        else:
            decisive_scores = ncdc_scores
            metric_used = 'NCDC'
            optimal_k = find_optimal_k(ncdc_scores, 'NCDC')
    else: # SC Only Analysis
        decisive_scores = sc_scores
        metric_used = 'SC'
        optimal_k = find_optimal_k(sc_scores, 'SC')

    kmeans = KMeans(n_clusters=optimal_k, random_state=random_seed, n_init=10).fit(data_to_cluster)
    df_2025['State_ID'] = kmeans.labels_
    centroids = kmeans.cluster_centers_.flatten()

    # Map state IDs to meaningful names based on optimal_k and sorted centroids
    if optimal_k > 1:
        sorted_centroids_indices = np.argsort(centroids)
        if optimal_k == 2:
            labels_map = ['Bear', 'Bull']
        elif optimal_k == 3:
            labels_map = ['Bear', 'Neutral', 'Bull']
        elif optimal_k == 4:
            labels_map = ['Strong Bear', 'Bear', 'Bull', 'Strong Bull']
        else:
            labels_map = [f'State {i+1}' for i in range(optimal_k)]

        ordered_labels = [labels_map[i] for i in sorted_centroids_indices]
        state_map = {idx: label for idx, label in enumerate(ordered_labels)}
        df_2025['State'] = df_2025['State_ID'].map(state_map)
    else:
        df_2025['State'] = 'State 1'
        ordered_labels = ['State 1']

    # --- Plotting Section ---
    fig = plt.figure(figsize=(18, 14))
    gs = GridSpec(3, 2, figure=fig, height_ratios=[3, 1, 2])
    fig.suptitle(f'S&P 500 Analysis for {year} | Optimal K={optimal_k} ({metric_used})', fontsize=18)

    # Panel 1: Price Chart (Top)
    ax0 = fig.add_subplot(gs[0, :])
    ax0.plot(df_2025['Date'], df_2025['Close'], color='black', lw=1.5, label='S&P 500 Close Price')
    ax0.set_title('S&P 500 Time Series')
    ax0.set_ylabel('Close Price')
    ax0.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.setp(ax0.get_xticklabels(), visible=False)

    # Panel 2: Quantized State Bar (Middle)
    ax1 = fig.add_subplot(gs[1, :], sharex=ax0)
    cmap = plt.get_cmap('viridis', optimal_k)

    if optimal_k > 1:
        for i, state_name in enumerate(ordered_labels):
            state_data = df_2025[df_2025['State'] == state_name]
            ax1.bar(state_data['Date'], [1] * len(state_data), width=1, color=cmap(i / (optimal_k - 1)), label=state_name)
    else:
        ax1.bar(df_2025['Date'], [1] * len(df_2025), width=1, color=cmap(0), label='State 1')

    ax1.set_yticks([])
    ax1.set_ylim(0, 1)
    ax1.set_ylabel('Market State')
    ax1.grid(axis='x', which='both', linestyle='--', linewidth=0.5)
    ax1.invert_yaxis()

    # Panel 3: Metrics Plot (Bottom Left)
    ax2 = fig.add_subplot(gs[2, 0])
    ax2.plot(list(sc_scores.keys()), list(sc_scores.values()), marker='o', linestyle='-', color='tab:blue', label='SC Score')
    ax2.plot(list(ncdc_scores.keys()), list(ncdc_scores.values()), marker='x', linestyle='--', color='tab:green', label='NCDC Score')

    if analysis_type == 'SC+':
        if max_sc_for_decision > 0.65:
            ax2.plot(list(sc_scores.keys()), list(sc_scores.values()), marker='s', linestyle='-', lw=2.5, color='purple', label='SC+ (Decisive)')
        else:
            pass

    ax2.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal K = {optimal_k} ({metric_used})')
    ax2.set_title(f'Quantification Metrics ({analysis_type})')
    ax2.set_xlabel('Number of Clusters (K)')
    ax2.set_ylabel('Score')
    ax2.set_xticks(range(1, 9))
    ax2.axhline(y=0.65, color='gray', linestyle=':', label='SC Threshold (0.65)')
    ax2.legend()
    ax2.grid(True)

    # Panel 4: Boxplot Distribution (Bottom Right)
    ax3 = fig.add_subplot(gs[2, 1])
    if optimal_k > 1:
        sns.boxplot(ax=ax3, x='State_ID', y='Log_Returns', data=df_2025)
        ax3.set_xticklabels(ordered_labels, rotation=45, ha='right')
        ax3.set_xlabel('Identified State')
    else:
        sns.boxplot(ax=ax3, y='Log_Returns', data=df_2025)
        ax3.set_xlabel('Identified State')
        ax3.set_xticklabels(['State 1'])

    ax3.set_title('Distribution of Log Returns per State')
    ax3.set_ylabel('Log Returns')

    fig.tight_layout(rect=[0, 0, 1, 0.96])

    output_filename = f'Yearly_Analysis_{year}_{random_seed}_{analysis_type}.png'
    plt.savefig(output_filename)
    print(f"\nFigure saved as '{output_filename}'")
    plt.show()

    print(f"--- Results for Random Seed {random_seed} ({analysis_type}) ---")
    print(f"Optimal K: {optimal_k} via {metric_used}")
    print(f"Max SC Score (K>1): {max_sc_for_decision:.4f}")
    print(f"SC Scores: {sc_scores}")
    print(f"NCDC Scores: {ncdc_scores}")
    print("-" * 50)

# --- Main Script Execution ---
df_2025_full = df_full[df_full['Date'].dt.year == 2025].copy()
df_2025_full['Log_Returns'] = np.log(df_2025_full['Close'] / df_2025_full['Close'].shift(1))
df_2025_full.dropna(subset=['Log_Returns'], inplace=True)
data_to_cluster = df_2025_full['Log_Returns'].values.reshape(-1, 1)

# Perform SC+ analysis and plotting
for seed in RANDOM_SEEDS:
    perform_analysis_and_plot(data_to_cluster, 2025, seed, df_2025_full.copy(), analysis_type='SC+')

# Perform SC-only analysis and plotting
for seed in RANDOM_SEEDS:
    perform_analysis_and_plot(data_to_cluster, 2025, seed, df_2025_full.copy(), analysis_type='SC_ONLY')
```
