```python
#news2016zh_valid_df = pd.read_json('data/corpora/chinese/new2016zh/news2016zh_train.json',encoding = 'utf-8-sig', lines=True)
#len(news2016zh_valid_df)
#memory overrun, need line-by-line processing!
import json
with open('data/corpora/chinese/new2016zh/news2016zh_train.json') as file:
    with open('data/corpora/chinese/new2016zh/news2016zh_train.txt', 'w') as fout:
        while True:
            line = file.readline()
            if not line:
                break
            j = json.loads(line)
            #print('title',j['title'])
            #print('desc',j['desc'])
            #print('content',j['content'])
            fout.write(j['title'])
            fout.write('\n')    
            fout.write(j['desc'])
            fout.write('\n')    
            fout.write(j['content'])
            fout.write('\n')    

```
