```python
brown_text = url_text("http://www.sls.hawaii.edu/bley-vroman/brown_nolines.txt").lower()
print(len(brown_text))
```

```python
max_n = 7 # https://www.sciencedirect.com/science/article/abs/pii/0378375886901692

brown_counter = grams_init(max_n)
text_grams_count(brown_counter,brown_text,max_n)
```

```python
for i in range(len(brown_counter)):
    counter = brown_counter[i]
    plot_dict(counter,'gram','freq',str(i),20)
```

```python
#denominate 2grams
counts2 = {}
for gram in brown_counter[1]:
    a = gram[0]
    b = gram[1]
    counts2[gram] = brown_counter[1][gram] / (brown_counter[0][a] + brown_counter[0][b])
plot_dict(brown_counter[1],'gram','freq','',30)
plot_dict(counts2,'gram','freq','',30)
```

```python
#denominate 3grams by 2grams
counts3 = {}
for gram in brown_counter[2]:
    a = gram[0:2]
    b = gram[1:3]
    counts3[gram] = brown_counter[2][gram] / (brown_counter[1][a] + brown_counter[1][b])
plot_dict(brown_counter[2],'gram','freq','',30)
plot_dict(counts3,'gram','freq','',30)
```

```python
#denominate 3grams by 1grams
counts3 = {}
for gram in brown_counter[2]:
    a = gram[0]
    b = gram[1]
    c = gram[2]
    counts3[gram] = brown_counter[2][gram] / (brown_counter[0][a] + brown_counter[0][b] + brown_counter[0][c])
plot_dict(brown_counter[2],'gram','freq','',30)
plot_dict(counts3,'gram','freq','',30)
```
