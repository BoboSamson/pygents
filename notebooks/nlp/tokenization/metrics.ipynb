```python
import os, sys
cwd = os.getcwd()
project_path = cwd[:cwd.find('pygents')+7]
if project_path not in sys.path: sys.path.append(project_path)
os.chdir(project_path) 

import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,3)
```

```python
df = pd.DataFrame([1,2,3,4,5,20,7,8,9,10,11,12,13,20,15,16,17,18,19,20],columns=['f'])
df['df'] = df['f'].diff()
df['dvf'] = df['f'] -  df['f'].mean()
# “Peak value” defined in https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2655800/ as a value of derivative on previous 
# transition minus value of TF derivative on following transition, which outlines sharp positive extremums of the function. 
df['peak'] = df['df'] - df['df'].shift(-1) 
# It can be interpreted as negative second derivative (derivative of derivative) shifted one point back. 
df['ddf'] = -df['df'].diff().shift(-1)
df[['f','dvf']].plot()
df[['f','df']].plot()
df[['f','peak']].plot()
df[['f','ddf']].plot()
```
