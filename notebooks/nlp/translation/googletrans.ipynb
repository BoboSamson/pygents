```python
def text_lines(path,debug = False):
    with open(path,errors='ignore') as f:
        lines = f.readlines()
    return [line.strip() for line in lines]
```

```python
lines = text_lines('data/media/brown_nolines.txt')
print(len(lines))
lines = [l for l in lines if len(l) > 0]
print(len(lines))
```

```python
lines = text_lines('data/media/brown_nolines.txt')
print(len(lines))
lines = [l for l in lines if len(l) > 0]
print(len(lines))
```

```python
for l in lines[:10]:
    t = translator.translate(l,'ru')
    print(t.origin)
    print(t.text)
```

```python
ts = translator.translate(lines[:4],'ru')
for t in ts:
    print(t.origin)
    print(t.text)
    
```

```python
#https://stackoverflow.com/questions/732109/chinese-language-codes
#0x0404 zh-TW Chinese (Taiwan)                     0x7C04 zh-CHT Chinese (Traditional)
#0x0804 zh-CN Chinese (People's Republic of China) 0x0004 zh-CHS Chinese (Simplified)
#0x0C04 zh-HK Chinese (Hong Kong S.A.R.)           0x7C04 zh-CHT Chinese (Traditional)
#0x1004 zh-SG Chinese (Singapore)                  0x0004 zh-CHS Chinese (Simplified)
#0x1404 zh-MO Chinese (Macau S.A.R.)               0x7C04 zh-CHT Chinese (Traditional)

ts = translator.translate(lines[:4],'zh-cn')
for t in ts:
    print(t.origin)
    print(t.text)
```

```python
lines_ru = []
for l in lines:
    t = translator.translate(l,'ru')
    if len(t.text) == 0:
        print(cnt,t.origin,'->',t.text)
        break
    lines_ru.append(t.text)
    cnt = len(lines_ru)
    if cnt % 100 == 0:
        print(cnt,t.origin,'->',t.text)
```
