```python
import kagglehub
import pandas as pd
from pathlib import Path
import shutil
import os

def load_kaggle_dataset_to_cwd(dataset_identifier, filename):
    """Download a Kaggle dataset and load it into the current working directory"""
    
    # 1. Download dataset (returns path where Kaggle stored it)
    downloaded_path = Path(kagglehub.dataset_download(dataset_identifier))
    print(f"Dataset downloaded to Kaggle cache: {downloaded_path}")
    
    # 2. Prepare paths
    source_file = downloaded_path / filename
    dest_file = Path.cwd() / filename
    
    # 3. Copy to current working directory
    if source_file.exists():
        shutil.copy2(source_file, dest_file)
        print(f"Copied dataset to current directory: {dest_file}")
        
        # 4. Load into pandas
        df = pd.read_csv(dest_file)
        print(f"Successfully loaded dataset with shape: {df.shape}")
        return df
    else:
        available_files = [f.name for f in downloaded_path.glob('*')]
        raise FileNotFoundError(
            f"File '{filename}' not found in dataset. Available files: {available_files}"
        )

# Usage
try:
    df = load_kaggle_dataset_to_cwd(
        dataset_identifier="sagarikashreevastava/cognitive-distortion-detetction-dataset",
        filename="Annotated_data.csv"
    )
    print(df.head())
except Exception as e:
    print(f"Error: {e}")
```
